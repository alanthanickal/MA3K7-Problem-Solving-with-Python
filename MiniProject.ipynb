```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    if x % 2 == 1:  # Only define g(x) for odd integers
        sum_result = 0
        for i in range(1, (x + 1)//2 + 1):
            sum_result += 1 / (2*i - 1)
        return sum_result

def h(x):
    if x % 2 == 0:  # Only define h(x) for even integers
        sum_result = 0
        for i in range(1, x//2 + 1):
            sum_result += 1 / (2*i)
        return sum_result

x_values = np.arange(1, 63)  # Adjust the range of x values as needed
y_values_g = [g(x) for x in x_values]
y_values_h = [h(x) for x in x_values]

x_values_g = [x for x, y in zip(x_values, y_values_g) if y is not None]
x_values_h = [x for x, y in zip(x_values, y_values_h) if y is not None]
y_values_g = [y for y in y_values_g if y is not None]
y_values_h = [y for y in y_values_h if y is not None]

plt.figure(figsize=(15, 6))  # Set figure size to width=10 inches, height=6 inches
plt.plot(x_values_g, y_values_g, 'o', label='g(n)')
plt.plot(x_values_h, y_values_h, 's', label='h(n)')
plt.xticks(np.arange(1, 63, 1))  # Set x-axis tick locations to integers from 1 to 20
plt.xlabel('n')
plt.ylabel('2f(n)')
plt.title('Plot of 2f(n) for odd and even n')
plt.legend()
plt.grid(True)
plt.show()

print(g(13))
print(g(15))
print(h(60))
print(h(62))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of values for n
n_values = np.linspace(0.1, 41, 400)

# Define the functions
def f(n):
    return (n - 3) / n

def g(n):
    return np.full_like(n, 0.5)

# Plot the graphs
plt.plot(n_values, f(n_values), label='y = (n - 3) / n')
plt.plot(n_values, g(n_values), label='y = 0.5')

# Add labels and title
plt.xlabel('n')
plt.ylabel('y')
plt.ylim(-2, 2)
plt.xlim(0,40)
plt.title('')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
def f(n):
    return (n // 4) + 1 if n % 4 == 0 else (n / 4) + 0.5

def play_game(n):
    # Initialize the array with values of 0
    glasses = [0] * n

    # Define the number of rounds
    num_rounds = f(n)
    
# Iterate through the rounds
    for _ in range(int(num_rounds)):  # Convert num_rounds to an integer
        # Ali's turn
        for i in range(0, n, 2):
            glasses[i] += 0.5 / (n // 2)
        
        # Beth's turn
        max_value = max(glasses)
        max_index = glasses.index(max_value)
        glasses[max_index] = 0
        
        # Empty one more glass
        if max_index - 1 == -1:
            glasses[n - 1] = 0  # Set index to n - 1
        else:
            glasses[max_index - 1] = 0  # Empty the glass to the left
        
        if max_index + 1 == n:
            glasses[0] = 0  # Set index to 0
        else:
            glasses[max_index + 1] = 0  # Empty the glass to the right
    
    # Print the final values in each array
    for i, val in enumerate(glasses):
        print(f"[{i}] = {val}")

# Test the game with any even n greater than or equal to 8
play_game(8)
```
