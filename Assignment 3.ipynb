```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.optimize import minimize

# Define the initial list of numbers
n = 2024 # n is the largest number in the hat
numbers = list(range(1, n + 1))

def compute_difference(numbers):
    # Make a copy of the list
    numbers_copy = numbers[:]
   
    # Randomly select two numbers from the list
    index1 = random.randint(0, len(numbers_copy) - 1)
    index2 = random.randint(0, len(numbers_copy) - 1)
    while index2 == index1:  # Ensure index2 is different from index1
        index2 = random.randint(0, len(numbers_copy) - 1)
   
    # Compute the difference between the selected numbers
    diff = abs(numbers_copy[index1] - numbers_copy[index2])
   
    # Remove the selected numbers from the list
    max_index = max(index1, index2)
    min_index = min(index1, index2)
    numbers.pop(max_index)
    numbers.pop(min_index)  # Adjust index after removing max_index
    if min_index < max_index:
        max_index -= 1
   
    # Add the difference to the list
    numbers.append(diff)    
    return numbers

# Specify the number of times to play the game
p = 10000

def play_game(numbers, p):
    final_numbers = []
    for i in range(p):
        numbers_copy = numbers[:]
        while len(numbers_copy) > 1:
            numbers_copy = compute_difference(numbers_copy)
        final_numbers.append(numbers_copy[0])
    return final_numbers

# Play the game n times and print the final numbers
final_numbers = play_game(numbers, p)
print(f"Final numbers after playing the game {p} times:", final_numbers)

# Define the histogram
bin_edges = np.arange(min(final_numbers) - 0.5, max(final_numbers) + 1.5, 1)
plt.figure(figsize=(20, 10))  # Set the figure size to a large value
plt.hist(final_numbers, bins=bin_edges, edgecolor='black', density=True, label='Histogram')
plt.xlabel('Final Number')
plt.ylabel('Density')
plt.title('Histogram of Final Numbers')

# Calculate the sample mean and variance of the histogram data
sample_mean = np.mean(final_numbers)
sample_var = np.var(final_numbers)

# Add labels, legend, and grid, and plotting
plt.xlabel('Final Number')
plt.xticks(np.arange(0, max(final_numbers) + 1, 2))
plt.xlim(min(final_numbers) - 0.5, max(final_numbers) + 0.5)
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()

def count_numbers(final_numbers):
    count_dict = {}  # To store counts of each number
    total_numbers = len(final_numbers)
   
    for num in final_numbers:
        if num in count_dict:
            count_dict[num] += 1
        else:
            count_dict[num] = 1

# Sort the dictionary based on counts in descending order
    sorted_counts = {k: v for k, v in sorted(count_dict.items(), key=lambda item: item[1], reverse=True)}

    
    percentage_dict = {num: (count / total_numbers) * 100 for num, count in sorted_counts.items()}
   
    return sorted_counts, percentage_dict

# Count the numbers and calculate percentages
counts, percentages = count_numbers(final_numbers)

# Print counts and percentages
print("Counts and percentages of each number (in descending order):")
for num, count in counts.items():
    print(f"{num}: Count = {count}, Percentage = {percentages[num]:.2f}%")
```
